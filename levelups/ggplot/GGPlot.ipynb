##GGPLOT
http://ggplot.yhathq.com/docs/index.html

Look at meat data

```python
meat.dtypes
```

```python
meat = meat.fillna(0)
```

GGplot works on layers.
Start with creating the plot using an aestetics object

```python
plot = ggplot(aes(x='date', y='veal'), data=meat) 
```

```python
plot + geom_point()+ geom_rect(xmin=meat.date.min(), xmax=meat.date.max(), ymin=0, ymax=100)
```

```python
plot + geom_point() + ggtitle("Veal prices")
```

```python
plot + geom_point() + stat_smooth(colour='red', span=0.9) 
```

```python
ggplot(aes(x='date', y='beef'), data=meat) +\
    geom_point() +\
    stat_smooth(colour='blue', span=0.2)
```

```python
ggplot(aes(x='date', y='beef'), data=meat) +\
    geom_point() + scale_y_log(base=2)
```

```python
ggplot(aes(x='date', y='beef'), data=meat) +\
    geom_point() + scale_y_log(base=2) + scale_x_reverse()
```

#Diamonds

Let's look at diamond data. Show distributions of color

```python
diamonds.head()
```

```python
plot = ggplot(aes(x='price', fill='color'), data=diamonds) 
```

```python
plot + geom_histogram()
```

Facetting is a powerful tool to show how different factors influence price and color

```python
ggplot(diamonds, aes(x='price', fill='color')) +\
    geom_density(alpha=0.25) +\
    facet_wrap("cut")
```

Histograms

```python
ggplot(diamonds, aes(x='price', fill='color')) +\
    geom_histogram(color='white')
```

Scatterplots are also easy

```python
plot = ggplot(aes(x='carat', y='price', color='color'), data=diamonds)
```

```python
plot + geom_point(alpha=0.5)
```

#Melted meat

Melt the data to use multi line plots

```python
meat_lng = pd.melt(meat, id_vars=['date'])
print meat_lng.head (2)
```

```python
meat_lng.head()
```

```python
plot = ggplot(aes(x='date', y='value', color='variable'), data=meat_lng)
```

```python
plot + geom_line(alpha=0.8) 
```

```python
mydata = pd.DataFrame({'x':np.arange(10)})

ggplot(aes(x='x'), data=mydata) + \
    stat_function(fun=np.sin, color="red") + \
    stat_function(fun=np.cos, color="blue")
```
